```python
import rasterio
import rasterio.plot
import pyproj
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
redband = 'cred.tif'
nirband = 'cnir.tif'

with rasterio.open('cred.tif') as src:
    red=src.read(1)
    
    
plt.imshow(red)
plt.colorbar()
plt.title('{}\nRed {}'.format(redband, red.shape))
plt.show()
with rasterio.open('cnir.tif') as src:
    nir = src.read(1)

plt.imshow(nir)
plt.colorbar()
plt.title('{}\nNIR {}'.format(nirband, nir.shape))
plt.xlabel('Column #')
plt.ylabel('Row #')
plt.show()

def calc_ndvi(nir,red):
    '''Calculate NDVI from integer arrays'''
    nir = nir.astype('f4')
    red = red.astype('f4')
    ndvi =  ((nir- red) / (nir + red)).astype(np.float32)
    return ndvi

ndvi = calc_ndvi(nir,red)
plt.imshow(ndvi, cmap='RdYlGn',vmin=-1,vmax=1)
plt.show()
```

```python
import rasterio
import rasterio.plot
import pyproj
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
redband = 'cred.tif'
nirband = 'cnir.tif'
blueband='cblue.tif'
with rasterio.open('cred.tif') as src:
    red=src.read(1)
plt.imshow(red)
plt.colorbar()
plt.title('{}\nRed {}'.format(redband, red.shape))
plt.show()
with rasterio.open('cnir.tif') as src:
    nir = src.read(1)
plt.imshow(nir)
plt.colorbar()
plt.title('{}\nNIR {}'.format(nirband, nir.shape))
plt.xlabel('Column #')
plt.ylabel('Row #')
plt.show()
with rasterio.open('cblue.tif') as src:
    blue = src.read(1)

plt.imshow(blue)
plt.colorbar()
plt.title('{}\nBLUE {}'.format(blueband, blue.shape))
plt.xlabel('Column #')
plt.ylabel('Row #')
plt.show()

def calc_evi(nir,red,blue):
    '''Calculate EVI from integer arrays'''
    nir = nir.astype('f4')
    red = red.astype('f4')
    blue = blue.astype('f4')
    numerator = nir-red
    c1 = red
    c2 = blue
    c1 = c1+nir
    denominator = c1-c2
    result = numerator/denominator
    result = result
    return 2*result

evi = calc_evi(nir,red,blue)
plt.imshow(evi, cmap='RdBu',vmin=-1,vmax=1)

plt.savefig('new.png')
plt.show()
```

```python
import rasterio
import rasterio.plot
import pyproj
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
redband = 'cred.tif'
nirband = 'cnir.tif'

with rasterio.open('cred.tif') as src:
    red=src.read(1)
plt.imshow(red)
plt.colorbar()
plt.title('{}\nRed {}'.format(redband, red.shape))
plt.show()
with rasterio.open('cnir.tif') as src:
    nir = src.read(1)

plt.imshow(nir)
plt.colorbar()
plt.title('{}\nNIR {}'.format(nirband, nir.shape))
plt.xlabel('Column #')
plt.ylabel('Row #')
plt.show()

def calc_savi(nir,red):
    '''Calculate SAVI from integer arrays'''
    nir = nir.astype('f4')
    red = red.astype('f4')
    ndvi =  ((nir- red) / (nir + red+0.5))*1.5
    return ndvi

savi = calc_savi(nir,red)
plt.imshow(savi, cmap='RdYlGn',vmin=-1,vmax=1)
plt.show()
```
